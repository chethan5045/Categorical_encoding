```python
# Task: Categorical Encoding + Feature Scaling

import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler

# Dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

# Task 1 — Categorical Encoding


# Gender → Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# City → One-Hot Encoding
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# Size → Ordered Encoding
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)
# Task 2 — Feature Scaling


# Using RobustScaler due to outlier in Fare (450)
scaler = RobustScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

# Output
df
print(df)
```

Reason for using RobustScaler:
The Fare column contains an outlier value (450). RobustScaler uses median and IQR, making it resistant to outliers, unlike StandardScaler or MinMaxScaler.
